```python
import pandas as pd
import re
df = pd.read_excel("/Users/mach/Desktop/H1.xls")

# Function to manually parse and separate concatenated URLs
def manual_parse_urls(text):
    # Find all occurrences of "https://"
    starts = [match.start() for match in re.finditer(r'https://', text)]

    urls = []
    for i in range(len(starts)):
        # Extract URL from the current "https://" to the next one or end of string
        start = starts[i]
        end = starts[i + 1] if i + 1 < len(starts) else None
        url = text[start:end].strip()
        urls.append(url)
    return "\n".join(urls) if urls else None

# Apply the manual parsing function to the DataFrame
df['Manually Extracted URLs'] = df['Online Document Address （必填）'].apply(manual_parse_urls)

# Dropping rows where no URLs were extracted
df_manually_extracted = df.dropna(subset=['Manually Extracted URLs'])

# Saving the data with manually extracted URLs to a new Excel file
output_manually_extracted_file_xlsx = '/Users/mach/Desktop/New_H1.xlsx'
df_manually_extracted[['Manually Extracted URLs']].to_excel(output_manually_extracted_file_xlsx, index=False)

output_manually_extracted_file_xlsx

df_manually_extracted[['Manually Extracted URLs']]
```
